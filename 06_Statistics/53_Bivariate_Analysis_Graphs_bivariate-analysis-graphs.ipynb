### 1. Categorical vs Categorical


```python
import seaborn as sns
data = pd.DataFrame({
    'gender': ['M','F','M','F','M','F','M','F'],
    'churn': [0,1,0,1,0,1,0,1],
    'segment': ['A','B','A','C','B','A','C','B']
})

# Stacked Bar Chart
pd.crosstab(data['gender'], data['churn']).plot(kind='bar', stacked=True)
plt.title('Churn by Gender')
plt.ylabel('Count')
plt.legend(title='Churn')
plt.show()
```

### 2. Numerical vs Numerical


```python
import numpy as np
np.random.seed(42)
age = np.random.normal(35, 10, 100)
sales = 200 + 5*age + np.random.normal(0, 50, 100)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(age, sales, alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Sales')
plt.title('Age vs Sales')

plt.subplot(1,2,2)
corr = np.corrcoef(age, sales)[0,1]
print(f"Pearson Correlation: {corr:.3f}")
sns.regplot(x=age, y=sales)
plt.title(f'Sales ~ Age (r={corr:.3f})')
plt.tight_layout()
plt.show()
```

### 3. Categorical vs Numerical


```python
data_bi = pd.DataFrame({
    'category': ['A','A','B','B','C','C','A','B','C','A'],
    'sales': [250, 320, 180, 290, 350, 420, 150, 310, 280, 340]
})

plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
sns.boxplot(x='category', y='sales', data=data_bi)
plt.title('Sales Distribution by Category')

plt.subplot(1,2,2)
sns.violinplot(x='category', y='sales', data=data_bi)
plt.title('Sales Density by Category')

plt.tight_layout()
plt.show()
```
