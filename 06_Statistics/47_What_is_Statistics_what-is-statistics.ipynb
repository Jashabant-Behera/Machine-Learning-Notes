## What is Statistics? {\#what-is-statistics}

### Definition

**Statistics** is the science of **collecting, analyzing, interpreting, and presenting data** to make informed decisions.

### Domains \& Use Cases

| Domain | Use Case Example |
| :-- | :-- |
| **Machine Learning** | Model evaluation, feature selection |
| **Healthcare** | Clinical trial analysis, survival rates |
| **Finance** | Risk assessment, portfolio optimization |
| **Marketing** | Customer segmentation, A/B testing |
| **Manufacturing** | Quality control, predictive maintenance |

### ML/DL Context

```
Statistics provides the foundation for:
- Model validation (p-values, confidence intervals)
- Feature importance analysis
- Overfitting detection
- A/B testing results interpretation
```

**Example**: Test if new ML algorithm improves accuracy from 85% to 90%


````python
# Quick Example
import numpy as np
old_accuracy = [0.85, 0.87, 0.84, 0.86]
new_accuracy = [0.89, 0.91, 0.88, 0.90]
print(f"Old Mean: {np.mean(old_accuracy):.3f}")
print(f"New Mean: {np.mean(new_accuracy):.3f}")
# Output: Old Mean: 0.855, New Mean: 0.895
````
