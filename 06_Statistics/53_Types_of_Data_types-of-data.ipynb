## Types of Data {\#types-of-data}

### 1. Categorical Data

```
Nominal: No order (gender, city, product_category)
Ordinal: Ordered (ratings: poor/good/excellent)
```


### 2. Numerical Data

```
Discrete: Countable (number of items, visits)
Continuous: Measurable (height, temperature, sales)
```

**Code Example**:


````python
import pandas as pd
data = pd.DataFrame({
    'gender': ['M','F','M','F'],      # Nominal
    'rating': [1,5,3,4],             # Ordinal
    'visits': [2,0,5,1],            # Discrete
    'sales': [250.5, 320.0, 180.2, 290.8]  # Continuous
})
print(data.dtypes)
````
