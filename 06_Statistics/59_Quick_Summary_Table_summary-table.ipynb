## Quick Summary Table {\#summary-table}

| Concept | ML Use Case | Formula | Best For |
| :-- | :-- | :-- | :-- |
| **Mean** | Feature scaling | $\bar{x}$ | Symmetric data |
| **Median** | Outlier resistant | Middle value | Skewed data |
| **Mode** | Imputation | Most frequent | Categorical |
| **SD** | Z-score norm | $\sqrt{Variance}$ | Model stability |
| **CV** | Scale comparison | $\sigma/\bar{x}$ | Different units |
| **t-test** | A/B testing | t-statistic | 2 groups |
| **ANOVA** | Model comparison | F-statistic | 3+ groups |
| **Chi2** | Feature selection | $\chi^2$ | Categorical |
