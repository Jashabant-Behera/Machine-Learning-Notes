## 2. Why Tensors Matter in Machine Learning

### How Tensors Flow Through a Neural Network

```
Input Data (Tensor)
        |
        v
    Neural Network Layer 1
        |
        v
    Transformation (Matrix Multiplication)
        |
        v
    Neural Network Layer 2
        |
        v
    Output Prediction (Tensor)

Example:
Input Image: 3D Tensor (Height x Width x Color Channels)
        |
        v
Flatten to 1D Vector (Input Tensor)
        |
        v
Matrix Multiply with Weights (2D Tensor)
        |
        v
Output: Probability Distribution (1D Tensor)
```

### Why We Need Tensors

1. **Efficient Computation**: GPU operations are optimized for multi-dimensional arrays
2. **Natural Representation**: Images, videos, sequences naturally fit tensor structure
3. **Batch Processing**: Handle multiple examples simultaneously (batch dimension)
4. **Deep Learning**: Neural networks operate on tensors
