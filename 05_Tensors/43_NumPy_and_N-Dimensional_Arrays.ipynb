## 4. NumPy and N-Dimensional Arrays

### Understanding numpy.ndim

`ndim` returns the number of dimensions (axes) in the array.


```python
import numpy as np

# 0D tensor (scalar)
tensor_0d = np.array(42)
print(f"ndim: {tensor_0d.ndim}")        # Output: 0

# 1D tensor (vector)
tensor_1d = np.array([1, 2, 3, 4, 5])
print(f"ndim: {tensor_1d.ndim}")        # Output: 1

# 2D tensor (matrix)
tensor_2d = np.array([[1, 2, 3], [4, 5, 6]])
print(f"ndim: {tensor_2d.ndim}")        # Output: 2

# 3D tensor (cube)
tensor_3d = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(f"ndim: {tensor_3d.ndim}")        # Output: 3

# 4D tensor (hypercube)
tensor_4d = np.random.rand(2, 3, 4, 5)
print(f"ndim: {tensor_4d.ndim}")        # Output: 4

# 5D tensor
tensor_5d = np.random.rand(2, 3, 4, 5, 6)
print(f"ndim: {tensor_5d.ndim}")        # Output: 5
```
