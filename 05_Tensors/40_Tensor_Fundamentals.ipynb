## 3. Tensor Fundamentals

### Key Tensor Properties

**1. Dimensionality (Rank)**
- Number of axes/dimensions in the tensor
- 0D tensor = rank 0 (scalar)
- 1D tensor = rank 1 (vector)
- 2D tensor = rank 2 (matrix)
- ND tensor = rank N

**2. Shape**
- Size along each dimension
- Example: Shape (32, 28, 28, 3) means:
  - 32 samples
  - 28 pixels height
  - 28 pixels width
  - 3 color channels

**3. Data Type (dtype)**
- What kind of numbers: int32, float32, float64, etc.
- Important for memory efficiency and precision

**4. Data**
- The actual numerical values

### NumPy: Python's Tensor Library

NumPy (Numerical Python) is the fundamental package for multi-dimensional arrays.


```python
import numpy as np

# Create tensors
scalar = np.array(42)
vector = np.array([1, 2, 3])
matrix = np.array([[1, 2], [3, 4]])
cube = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Inspect properties
print(vector.ndim)        # Number of dimensions (rank)
print(vector.shape)       # Shape along each axis
print(vector.dtype)       # Data type
print(vector.size)        # Total number of elements
```
