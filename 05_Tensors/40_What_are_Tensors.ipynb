## 1. What are Tensors?

### Simple Definition

A **tensor** is a multi-dimensional array of numbers. It's a generalization of:
- Scalars (single number)
- Vectors (list of numbers)
- Matrices (grid of numbers)
- Higher-dimensional arrays (cube, hypercube, etc.)

### Why the Name "Tensor"?

In physics, a tensor is a mathematical object that describes physical properties and how they transform under coordinate changes. In machine learning, we use the term more loosely to mean any multi-dimensional array.

### Tensor vs Array vs Matrix

```
Term              Definition                          Example
----              ----------                          -------
Scalar (0D)       Single number                       42, 3.14
Vector (1D)       List of numbers                     [1, 2, 3, 4, 5]
Matrix (2D)       Grid of numbers                     [[1,2],[3,4]]
Tensor (ND)       Multi-dimensional array            Any N-dimensional structure
```
